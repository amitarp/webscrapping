```python
from bs4 import BeautifulSoup
import requests,openpyxl
import pandas
```

```python
source=requests.get("https://en.wikipedia.org/wiki/List_of_Harry_Potter_characters")
```

```python
#this is used to check if the url is valid or not
source.raise_for_status()
```

```python
soup=BeautifulSoup(source.text,'html.parser')
```

```python
characters=soup.find_all('a',class_='mw-redirect')
```

```python
characters
```

```python
for character in characters:
    print(character.text)
```
