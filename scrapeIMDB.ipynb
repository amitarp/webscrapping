```python
from bs4 import BeautifulSoup
import requests,openpyxl
import pandas
```

```python
source=requests.get("https://www.imdb.com/chart/top/")
```

```python
#this is used to check if the url is valid or not
source.raise_for_status()
```

```python
soup=BeautifulSoup(source.text,'html.parser')
```

```python
movies=soup.find('tbody',class_='lister-list').find_all('tr')
```

```python
len(movies)
```

```python
lis=[]
```

```python
for movie in movies:
    d={}
    d['Name']=movie.find('td',class_='titleColumn').a.text
    d['Rank']=movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
    d['Year']=movie.find('td',class_='titleColumn').span.text.strip('()')
    d['Rating']=movie.find('td', class_='ratingColumn imdbRating').strong.text
    lis.append(d)
   
```

```python
lis
```

```python
import pandas
df=pandas.DataFrame(lis)
df
```

```python
df.to_csv("MovieRating.csv")
```
